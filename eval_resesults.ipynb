### Experiment Parameters

Below is the explanation of the configurable parameters used to define the experiments. Note that boolean values accept various formats (e.g., `1/0`, `True/False`, `yes/no`, `y/n`, `t/f`).

| Parameter | Description | Accepted Values | Default | Notes & Constraints |
| :--- | :--- | :--- | :--- | :--- |
| <nobr>**`--bench`**</nobr> | Benchmark name. | `bbq`, `crows_pairs`, `discrim_eval`, `dt_toxic`, `mmlu`, `reddit_bias`, `stereo_set`, `toxic_ratings`, `wino_bias` | | Required during initialization (`init_bench=True`). |
| <nobr>**`--category`**</nobr> | Demographic category. | *String* (e.g., `"gender"`, `"race"`) | `None` | Depends on the specific benchmark. |
| <nobr>**`--is_cot`**</nobr> | Use chain-of-thought prompting. | *Boolean* (e.g., `0/1`, `True/False`) | `True` | Triggers a warning if `--gen_length` is < 1000. Automatically set to `False` if `--is_prob=True`. |
| <nobr>**`--is_third`**</nobr> | Allow a third neutral option. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | |
| <nobr>**`--is_prompt_var`**</nobr> | Prompt model with 54 prompt variations. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | |
| <nobr>**`--is_prob`**</nobr> | Evaluation based on option token probability distribution. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | **Note:** If `True`, automatically forces `--is_cot=False`, `--gen_length=None`, and `--n_gens=None`. |
| <nobr>**`--is_rand`**</nobr> | Enforce random answer. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | |
| <nobr>**`--is_chat`**</nobr> | Use chat templates. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | |
| <nobr>**`--gen_length`**</nobr> | Maximum length of generated text. | *Integer* (e.g., `512`, `1024`) | `2000` | Automatically disabled (set to `None`) if `--is_prob=True`. |
| <nobr>**`--n_gens`**</nobr> | Number of generations for each prompt. | *Integer* (e.g., `1`, `5`) | `25` | Automatically disabled (set to `None`) if `--is_prob=True`. |
| <nobr>**`--is_disamb`**</nobr> | Use disambiguated split of BBQ benchmark. | *Boolean* (e.g., `0/1`, `True/False`) | `False` | |
| <nobr>**`--subset_size`**</nobr> | Maximum number of samples to evaluate from the benchmark. | *Integer* (e.g., `100`, `1000`) | `None` | Useful for testing or debugging on a subset of the dataset. |
